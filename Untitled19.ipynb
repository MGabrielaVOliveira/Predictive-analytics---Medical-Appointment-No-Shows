```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#Analyzing dataset

df = pd.read_csv('KaggleV2-May-2016.csv')
df.head()
```

```python
# Checking for null
df.isnull().sum()
```

```python
# Transforming Columns string in data

df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
df['WaitingDays'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days
```

```python
# Transforming significant columns in 0 or 1

df['No-show-Cod'] = df['No-show'].map({'No': 0, 'Yes': 1})
df['Gender-Cod'] = df['Gender'].map({'F': 0, 'M': 1})
```

```python
df.head()
```

```python
sns.countplot(x='Gender', hue='No-show', data=df)
plt.show()
```

```python
sns.boxplot(x='No-show', y='Age', data=df)
plt.show()
```

```python
#Creating features for X

Features = ['Age','Scholarship','Gender-Cod','Hipertension','Diabetes','Alcoholism','Handcap','SMS_received','WaitingDays']
```

```python
X = df[Features]
Y = df['No-show-Cod']
```

```python
#Training
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#In this part, we have the predictive

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
#Accuracy visualization
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Visualization in heatmap
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
plt.show()
```

```python
# Positive results increase no-show rate at doctor's office.

coef = pd.DataFrame({'Feature': X.columns,'Coefficient': model.coef_[0]}).sort_values(by='Coefficient', ascending=False)
coef
```
